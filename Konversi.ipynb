```python
import os
import tensorflow as tf
from tensorflowjs.converters import tf_saved_model_conversion_v2

path_saved_model = "/content/drive/My Drive/Projek/saved_model"
path_output_tfjs = "/content/drive/My Drive/Projek/tfjs_model"

if os.path.exists(path_saved_model):
    try:
        tf_saved_model_conversion_v2.convert_tf_saved_model(
            saved_model_dir=path_saved_model,
            output_dir=path_output_tfjs
        )
        print("Berhasil")

    except Exception as e:
        print(f"Terjadi kesalahan: {e}")
```

```python
path_target_tflite = "/content/drive/My Drive/Projek/tflite_model.tflite"
path_label_drive = "/content/drive/My Drive/Projek/labels.txt"

#Buat label.txt
labels = ["Non Demented", "Mild Demented", "Very Mild Demented", "Moderate Demented"]

with open(path_label_drive, "w") as f:
    for kelas in labels:
        f.write(kelas + "\n")

#Konversi TF-Lite
if os.path.exists(path_saved_model):
    try:
        converter = tf.lite.TFLiteConverter.from_saved_model(path_saved_model)
        converter.optimizations = [tf.lite.Optimize.DEFAULT]
        tflite_model = converter.convert()

        with open(path_target_tflite, 'wb') as f:
            f.write(tflite_model)

        print("Berhasil")
    except Exception as e:
        print(f"Terjadi kesalahan: {e}")
```
